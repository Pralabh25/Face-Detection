```python
from google.colab import drive
drive.mount('/content/drive')
import os
import cv2
import dlib
import matplotlib.pyplot as plt
```

```python

root=("/content/drive/My Drive/pics")
os.chdir(root)
```

```python

hog_face_detector = dlib.get_frontal_face_detector()
# hog face detection model
```

```python

def hogDetectFace(image, hog_face_detector, display=True):
  '''
  Function: To detect the face using dlib
  Arguments:
  1:image=>the image where the person's face  has to be detected
  2:hog_face_detector=>model
  3:display=>if its true than input image with the bounding box is displayed
  Returns:
  1:output_image: a copy of input image with bounding box
  2:result:the output of face recognisition process
  '''
  height , width, _ = image.shape

  output_image = image.copy()

  img_in_RGB = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

  result = hog_face_detector(img_in_RGB,0)

  for boundingBox in result:
    x1 = boundingBox.left()

    y1=boundingBox.top()

    x2 = boundingBox.right()

    y2 = boundingBox.bottom()

    cv2.rectangle(output_image,pt1=(x1,y1),pt2=(x2,y2),color=(0,255,0),thickness=width//200)

    if display:
      plt.figure(figsize=[15,15])
      plt.subplot(121);plt.imshow(image[:,:,::-1]);plt.title("Original Image");plt.axis('off');
      plt.subplot(122);plt.imshow(output_image[:,:,::-1]);plt.title("Output Image");plt.axis('off');
    else:
      return output_image, result
```

```python

image=cv2.imread("/content/drive/My Drive/pics/image1.jpg")
hogDetectFace(image,hog_face_detector,display=True)
```
